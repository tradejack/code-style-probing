```python
import pandas as pd
import numpy as np
import os
fname_prefix = "/data/users/team2_capstone/code-style-probing/"
```

```python
clean_metrics_df = pd.read_csv(fname_prefix + "data/bigquery_metric_dedup_file_20220530/bigquery_metric_dedup_file0_20220530.csv")
os.listdir(fname_prefix + "data")
bq_metric_files = [
    #'bigquery_metric_dedup_file0_20220530.csv',
    #'py150k_metric_20220527.csv',
    'bigquery_metric_dedup_file_20220530/bigquery_metric_dedup_file1_20220530.csv',
    'bigquery_metric_dedup_file_20220530/bigquery_metric_dedup_file2_20220530.csv',
    'bigquery_metric_dedup_file_20220530/bigquery_metric_dedup_file3_20220530.csv',
    'bigquery_metric_dedup_file_20220530/bigquery_metric_dedup_file4_20220530.csv',
    'bigquery_metric_dedup_file_20220530/bigquery_metric_dedup_file5_20220530.csv',
    'bigquery_metric_dedup_file_20220530/bigquery_metric_dedup_file6_20220530.csv',
    'bigquery_metric_dedup_file_20220530/bigquery_metric_dedup_file7_20220530.csv',
    'bigquery_metric_dedup_file_20220530/bigquery_metric_dedup_file8_20220530.csv',
    'bigquery_metric_dedup_file_20220530/bigquery_metric_dedup_file9_20220530.csv' 
 ]

for file in bq_metric_files:
    temp_df = pd.read_csv(f'{fname_prefix}data/{file}')
    clean_metrics_df = pd.concat((clean_metrics_df , temp_df))
#os.listdir("data")
display(clean_metrics_df)
```

```python
orig_data = os.listdir(fname_prefix + "data/BigQuery/files")
orig_data_df = pd.DataFrame()
for file in orig_data:
    temp_df = pd.read_csv(f'{fname_prefix}data/BigQuery/files/{file}')
    orig_data_df = pd.concat((orig_data_df , temp_df))

display(orig_data_df)
```

```python
clean_metrics_df['file'] = [ x + y  for x, y in zip(clean_metrics_df['repository'], clean_metrics_df['filepath'])]
orig_data_df['file'] = [ x + y  for x, y in zip(orig_data_df['repository'], orig_data_df['filepath'])]
ex_files = list(clean_metrics_df['file'])
filtered_bq = orig_data_df[ ~orig_data_df['file'].isin(ex_files)]
train_df = orig_data_df[ orig_data_df['file'].isin(ex_files)]
display (filtered_bq) #393648
#display (orig_data_df)
```

```python
import hashlib
train_df = train_df[train_df['content'].notnull()]
filtered_bq = filtered_bq[filtered_bq['content'].notnull()]
filtered_bq = filtered_bq.reset_index(drop=True)
filtered_bq.at[0, 'content'] = list(train_df['content'][1])[0]
def checksum (source):
    return hashlib.md5(source.encode('utf-8')).hexdigest()
#print (checksum("apples"))
train_df['checksum'] =  train_df['content'].apply(checksum)
filtered_bq['checksum'] = filtered_bq['content'].apply(checksum)
display (filtered_bq)
display(train_df)
```

```python

orig_files = list(train_df['content'])
deduped_df = filtered_bq[ ~filtered_bq['content'].isin(orig_files)]
checksum_unique = deduped_df['content'].unique()
display (len(checksum_unique))
deduped_df.to_csv(fname_prefix + "data/evaluation_set.csv")
```
