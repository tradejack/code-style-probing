## Getting Repository Frequencies per Label

```python
from collections import Counter
import numpy as np
import math

num_classes = 26
counters = []

for i in range(0,num_classes):
    count = Counter()
    files = df[df.labels == str(i)]["file"]
    roots = []
    for file in files:
        if isinstance(file, str):
            path_parts = file.split("/")
            if path_parts[0] == "data":
                root = path_parts[1:3]
            else:
                root = path_parts[0:2]
            roots += ["/".join(root)]

    count.update(roots)
    counters += [count]
```

```python
def avg(x):
    sm = 0
    tot = 0
    for e in x:
        if isinstance(e, str):
            sm += float(e)
            tot += 1
    return sm/tot
avg(["1","2"])
```

```python
target_metrics = ["forks", "stars", "comment_avg_len", "comment_density", "ds_char_len_avg", "ds_density", "comprehensions", "snake_case_var_ratio", "lower_camel_case_var_ratio"]
metrics_dict = {key : [] for key in target_metrics}
tops = []
for e, count in enumerate(counters):
    print(f"Label: {e}, Support: {len(df[df.labels == str(e)])}")
    for t in target_metrics:
        av = avg(df[df.labels == str(e)][t])
        print(f"Average {t}: {av}")
        metrics_dict[t] += [av]

    tops += [count.most_common(10)]
    for c in count.most_common(10):
        print(c)
    #print(metrics_dict)
    print("\n------------------------\n")
```
